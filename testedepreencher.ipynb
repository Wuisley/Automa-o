```python
import pandas as pd
import pyautogui
import pyperclip
import time
import numpy
```

```python
#PREENCHENDO OS ÓRGÃOS
time.sleep(3)

for v in range(4):
    pyautogui.press('tab')

pyautogui.press('space')

for b in range(4):
    pyautogui.hotkey('shift', 'tab')

for i, dados in enumerate(df["CNPJ"]):
    cnpj = df.loc[i, "CNPJ"]
    nome = df.loc[i, "NOME"]
    competencia = df.loc[i, "COMPETENCIA"]
    poder = df.loc[i, "PODER"]
    tipo = df.loc[i, "TIPO"]

    pyperclip.copy(str(cnpj))
    pyautogui.hotkey('ctrl','v')
    pyautogui.press('Tab')

    pyperclip.copy(str(nome))
    pyautogui.hotkey('ctrl','v')
    pyautogui.press('Tab')

    pyperclip.copy(str(competencia))
    pyautogui.hotkey('ctrl','v')
    pyautogui.press('Tab')

    pyautogui.hotkey('alt', 'down')
    for n in range(poder):
        pyautogui.press('down')

    pyautogui.press('Tab')
      
    for t in range(tipo-1):
        pyautogui.press('down')
    
    pyautogui.press('Tab')
    pyautogui.press('enter')
        
        

    #pyautogui.PAUSE = 1
    
    pyautogui.hotkey('shift','tab')
    for t in range(tipo-1):
        pyautogui.press('left')
    pyautogui.hotkey('shift','tab')
    pyautogui.hotkey('shift','tab')
    pyautogui.hotkey('shift','tab')
    pyautogui.hotkey('shift','tab')
```
